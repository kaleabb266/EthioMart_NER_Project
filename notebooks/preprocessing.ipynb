```python
# Define a function to clean and tokenize the message column
def clean_and_tokenize_amharic_message(text):
    # Regular expression to match Amharic characters (\u1200-\u137F) and numbers (0-9)
    pattern = re.compile(r'[^\u1200-\u137F0-9]+')
    # Substitute the unwanted characters with an empty string
    cleaned_text = re.sub(pattern, ' ', str(text))
    # Tokenize by splitting based on spaces
    tokens = cleaned_text.split()
    # Join tokens with commas
    tokenized_text = ','.join(tokens)
    return tokenized_text

# Apply the cleaning and tokenizing function to the 'message' column
df['cleaned_tokenized_message'] = df['Message'].apply(clean_and_tokenize_amharic_message)
```

```python
# Function to check if a message contains only numeric tokens
def is_numeric_only(text):
    # Split the text by commas and spaces
    tokens = re.split(r'[,\s]+', text)
    # Check if all tokens are numeric
    return all(token.isdigit() for token in tokens if token)

# Apply the function to filter out rows where 'cleaned_tokenized_message' is numeric-only
df_filtered = df[~df['cleaned_tokenized_message'].apply(is_numeric_only)]
```
